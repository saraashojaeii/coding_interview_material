# Sets
## Concept

A set is like a dict with only keys. Great for membership tests and uniqueness.

## Key Python syntaxes
```python 
s = set()
s.add(x)
if x in s: ...
s.remove(x)   # KeyError if not found
s.discard(x)  # no error
```

### Template
```python
seen = set()
for x in arr:
    if x in seen:
        # we've seen duplicate x
        ...
    seen.add(x)

```


## Example
**Problem 1**: Contains duplicate?

````python
def contains_duplicate(nums: list[int]) -> bool:
    seen = set()
    for i in nums:
        if i in seen:
            return True
        seen.add(i)
    return False

print(contains_duplicate([1, 2, 3, 1]))  # True
print(contains_duplicate([1, 2, 3]))     # False
````

````python
from unittest import skip


def array_intersection(s1: list[int], s2: list[int]):
    unique_set = set()
    for i in s1:
        if i in unique_set:
            skip
        unique_set.add(i)
    for i in s2:
        if i in unique_set:
            skip
        unique_set.add(i)
    return unique_set

list1 = [1, 4, 8, 5]
list2 = [1, 7, 5, 3, 7, 9]
target = 9

print(array_intersection(list1, list2))
````

**Problem 3:** Missing Number

Input: array containing `n` distinct numbers from `0..n`, one missing

Output: the missing number.

````python

def missing_bumber(nums: list[int]):
    complete_set = set(range(0, len(nums)+1))
    print(complete_set)
    for i in complete_set:
        if i not in nums:
            return i
    

list1 = [1, 4, 3, 2]
list2 = [1, 4, 3, 2, 0, 5]

print(missing_bumber(list1))
print(missing_bumber(list2))
````
